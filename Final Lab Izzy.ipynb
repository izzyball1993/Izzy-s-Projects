REMINDER: To use Keras, we also require the relevant backend framework - TensorFlow. We will also import the relevant libraries to begin our work. 

```python
import pandas as pd
import numpy as np
import keras

import warnings
warnings.simplefilter('ignore', FutureWarning)
```

Now, we'll install the relevant data set and put it into a Pandas dataframe so that we can do the needful data analysis.

```python
filepath='https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DL0101EN/labs/data/concrete_data.csv'
concrete_data = pd.read_csv(filepath) #retrieving dataset from csv file and setting to dataframe

concrete_data.head() #printing the first few rows of the data set as a quick sanity check.
```

Quick sanity check for null values, dirty data etc.

```python
concrete_data.describe()
concrete_data.isnull().sum() 
```

Data must now be segmented into predictors and the target variable, which in this case will be the concrete sample strength.

```python
concrete_data_columns = concrete_data.columns
predictors = concrete_data[concrete_data_columns[concrete_data_columns != 'Strength']] #assign all cols except strength (target) to predictors
target = concrete_data['Strength'] # Strength column is our target as noted earlier
```

We'll now import the Keras packages required to successfully build our neural network

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from keras.models import Sequential
from keras.layers import Dense, Input
from keras.optimizers import Adam
```

PART A - Build a baseline model

We now begin the work of constructing our baseline neural network. Reminder of the requirements below:

- One hidden layer of 10 nodes, and a ReLU activation function

- Use the adam optimizer and the mean squared error  as the loss function.

1. Randomly split the data into a training and test sets by holding 30% of the data for testing. You can use the 
train_test_split
helper function from Scikit-learn.

2. Train the model on the training data using 50 epochs.

3. Evaluate the model on the test data and compute the mean squared error between the predicted concrete strength and the actual concrete strength. You can use the mean_squared_error function from Scikit-learn.

4. Repeat steps 1 - 3, 50 times, i.e., create a list of 50 mean squared errors.

5. Report the mean and the standard deviation of the mean squared errors.

```python
# define regression model
def regression_model():
    # create model
    model = Sequential()
    model.add(Input(shape=(predictors.shape[1],))) #we're going with un-normalized data for the time being (we'll normalize for part b
    model.add(Dense(10, activation='relu')) #one hidden layer of 10 nodes with ReLU activation function
    model.add(Dense(1))
    
    # compile model
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model

mse_list = [] #MSE storage list

for _ in range(50): #loop to give us 50 MSEs to analyze
    X_train, X_test, y_train, y_test = train_test_split(predictors, target, test_size=0.3, random_state=None)
    
    # Create and train the model
    model = regression_model()
    model.fit(X_train, y_train, epochs=50, verbose=0)
    
    # Predict and evaluate
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    mse_list.append(mse)

#now we calculate and report AVG and STD of the MSE's in our list
mean_mse = round(np.mean(mse_list),3)
std_mse = round(np.std(mse_list),3)

print(f'Mean MSE: {mean_mse}')
print(f'Standard Deviation of MSE: {std_mse}')
```

PART B - Normalize The Data

#Requirements: Repeat Part A but use a normalized version of the data. Recall that one way to normalize the data is by subtracting the mean from the individual predictors and dividing by the standard deviation.

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
from keras.models import Sequential
from keras.layers import Dense, Input


predictors = concrete_data[concrete_data_columns[concrete_data_columns != 'Strength']] #assign all cols except strength (target) to predictors
target = concrete_data['Strength'] # Strength column is our target as noted earlier
scaler = StandardScaler()
predictors_norm = scaler.fit_transform(predictors)
```

Once again, we will construct and run our regression model and observe differences. 

```python
# define normalized regression model
def norm_model(X, y):
    mse_list = []
    for _ in range(50):
        # Split the data into training and test sets
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=None)

        # Build the neural network model
        model = Sequential()
        model.add(Input(shape=(X_train.shape[1],)))  # Use Input layer here
        model.add(Dense(10, activation='relu')) 
        model.add(Dense(1))

        # Compile the model
        model.compile(optimizer='adam', loss='mean_squared_error')

        # Train the model
        model.fit(X_train, y_train, epochs=50, verbose=0)
        # Evaluate the model
        y_pred = model.predict(X_test)
        mse = mean_squared_error(y_test, y_pred)
        mse_list.append(mse)
    
    return mse_list

# Evaluate the model with normalized data
mse_list_norm = norm_model(predictors_norm, target)

#now we calculate and report AVG and STD of the MSE's in our list
mean_mse_norm = round(np.mean(mse_list_norm),3)
std_mse_norm = round(np.std(mse_list_norm),3)

print(f'Mean MSE: {mean_mse_norm}')
print(f'Standard Deviation of MSE: {std_mse_norm}')
```

Note above a decrease in average MSE and a significant decrease in STD of the MSE relative to exercise a)

PART C: Increase Number Of Epochs

Repeat Part B but use 100 epochs this time for training.

```python
def norm_model_100(X, y):
    mse_list = []
    for _ in range(100):
        # Split the data into training and test sets
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=None)

        # Build the neural network model
        model = Sequential()
        model.add(Input(shape=(X_train.shape[1],)))  # Use Input layer here
        model.add(Dense(10, activation='relu')) 
        model.add(Dense(1))

        # Compile the model
        model.compile(optimizer='adam', loss='mean_squared_error')

        # Train the model
        model.fit(X_train, y_train, epochs=100, verbose=0) #changing to 100 epochs
        # Evaluate the model
        y_pred = model.predict(X_test)
        mse = mean_squared_error(y_test, y_pred)
        mse_list.append(mse)
    
    return mse_list

# Evaluate the model with normalized data
mse_list_norm = norm_model_100(predictors_norm, target)

#now we calculate and report AVG and STD of the MSE's in our list
mean_mse_norm = round(np.mean(mse_list_norm),3)
std_mse_norm = round(np.std(mse_list_norm),3)

print(f'Mean MSE: {mean_mse_norm}')
print(f'Standard Deviation of MSE: {std_mse_norm}')
```

Relative to exercise b), we see significant reduction in average MSE and also in STD of the MSE running with 100 epochs instead of 50.

PART D: Increase The Number Of Hidden Layers

Requirements: Repeat part B but use a neural network with the following instead:

- Three hidden layers, each of 10 nodes and ReLU activation function.

```python
# define normalized regression model
def three_deep_model(X, y):
    mse_list = []
    for _ in range(50):
        # Split the data into training and test sets
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=None)

        # Build the neural network model
        model = Sequential()
        model.add(Input(shape=(X_train.shape[1],)))  # Use Input layer here
        model.add(Dense(10, activation='relu')) #1st hidden layer
        model.add(Dense(10, activation='relu')) #2nd hidden layer
        model.add(Dense(10, activation='relu')) #3rd hidden layer
        model.add(Dense(1))

        # Compile the model
        model.compile(optimizer='adam', loss='mean_squared_error')

        # Train the model
        model.fit(X_train, y_train, epochs=50, verbose=0) #back to 50 epochs as in exercise b)
        # Evaluate the model
        y_pred = model.predict(X_test)
        mse = mean_squared_error(y_test, y_pred)
        mse_list.append(mse)
    
    return mse_list

# Evaluate the model with normalized data
mse_list_norm = three_deep_model(predictors_norm, target)

#now we calculate and report AVG and STD of the MSE's in our list
mean_mse_norm = round(np.mean(mse_list_norm),3)
std_mse_norm = round(np.std(mse_list_norm),3)

print(f'Mean MSE: {mean_mse_norm}')
print(f'Standard Deviation of MSE: {std_mse_norm}')
```

Mean MSE and std of MSE is the lowest running the 3 layer deep netwrok in exercise D compared to exericse b, or any of the others. Note, run time per epoch is also slightly longer compared to the others.
